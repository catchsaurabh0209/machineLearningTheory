```python
import matplotlib.pyplot as plt # basically pyplot is responsible for visualization .
import numpy as np
x=np.arange(1,9,2)
print(x)
y=x+10
y1=2*x+12
# alpha define transparency.
plt.plot(x,y,linewidth=2,linestyle='-',color='b',alpha=1.0,marker='o') #bm First coordinate is placed on the horizontal axis. and  y on vertical axis.
plt.plot(x,y1,linewidth=2,linestyle='-',color='r',alpha=1.0,marker='o') #bm First coordinate is placed on the horizontal axis. and  y on vertical axis.

plt.title("Straight line graph")
plt.xlabel("Horizontal value")
plt.ylabel("vertical value")
plt.grid(True)
plt.legend(['line 1','line 2'],loc='best') #loc makes the the graph to adjust in best position and list containg list 1 and list 2 describe the the line the graph.
plt.show()

```

```python
plt.subplot(1,2,1)# the last coordinate(1) denotes the no. of the figure.
plt.plot(x,y)
plt.show()
```
